```python
# some chooses for travel modes

# car hire option
def car_hire():
  # some prompts
  print("\nYou have arrived at Enterprise office. One way daily rental cost is £80 + fuel cost. You can hire the car, drive for 8 hours and arrive in Edinburg today.")
  print("or, you may choose not to bother and come back another day for using the public transport.")
  print("What do you want to do? (1 or 2)")
  print("1). You don't bother with driving and come back another day for the public transport.")
  print("2). Go for Enterprise and hire a car for driving one way to Edinburg.")

  # take input()
  answer = input(">")
  
  if answer == "1":
    # the player is travelling another day!, call game_over() function with the "reason"
    game_over("Good choose! ...as there was an accident on M1 and it would have cost you 12 hours + £160 (£80 fuel, £80 rental cost) for driving. However, staying overnight in a hotel only costed you £100.")
  elif answer == "2":
    # the player travells today, with some unexpected costs...
    print("\nYou made it in Edinburg today, but it costs you 12 hours + £160 (£80 fuel, £80 rental cost) since there was a very serious accident on M1. One of them days... ")
    # activate play_again() function
    play_again()
  else:
    # call game_over() with "reason"
    game_over("Typed in the wrong way.")


# train mode option
def train_station():
  # give some prompts
  # '\n' is to print the line in a new line
  print("\nYou have arrived at the train station. There is distrubtion in all services for 8 hours due to signal issiues.")
  print("You can chose to travel another day")
  print("or, you can hire a car from Enterprise.")
  print("What would you do? (1 or 2)")
  print("1). Travel another day.")
  print("2). Hire a car.")

  # take input()
  answer = input(">")
  
  if answer == "1":
    # the player is travelling another day!
    game_over("You chosed to travel another day and stayed overnight in London with the hotel costing you £100.")
  elif answer == "2":
    # lead him to car hire option()
    print("\nYou will need to hire a car from Enterprise.")
    car_hire()
  else:
    # else call game_over() function with the "reason" argument
    game_over("Typed in the wrong way.")


# bus mode option
def bus_station():
  # some prompts
  # '\n' is to print the line in a new line
  print("\nYou have arrived at Victoria Bus Station.")
  print("One of them days! Drivers at strike for better pay. You can stay in London for one more night or chose to hire a car from enterprise. The cost of staying another night is £100 and one way rental cost £80 plus the fuel cost. Chose (1 or 2)")
  print("1). Stay another night.")
  print("2). Hire a car.")

  # take input()
  answer = input(">")

  if answer == "1":
    # the player travelling another day, call game_over() with "reason"
    game_over("You stayed another night in London and the hotel costed you £100.  ")
  elif answer == "2":
    # lead him to the car hire option()
    car_hire()
  else:
    # game_over() with "reason"
    game_over("Typed in the wrong way")


# function to ask play again or not
def play_again():
  print("\nDo you want to play again? (y or n)")
  
  # convert the player's input to lower_case
  answer = input(">").lower()

  if "y" in answer:
    # if player typed "yes" or "y" start the game from the beginning
    start()
  else:
    # if user types anything besides "yes" or "y", exit() the program
    exit()


# game_over function accepts an argument called "reason"
def game_over(reason):
  # print the "reason" in a new line (\n)
  print("\n" + reason)
  print("Game Over!")
  # ask player to play again or not by activating play_again() function
  play_again()


def start():
  # give some prompts.
  print("\nYou need to travel to Edinburg from London. You can chose to travel by Train, Bus or Hire a car.")
  print("Which travel mode you want to use? (train, bus or car hire)")
  
  # convert the player's input() to lower_case
  answer = input(">").lower()

  if "t" in answer:
    # if player typed "train" or "t" lead him to train_station()
    train_station()
  elif "b" in answer:
    # else if player typed "bus" or "b" lead him to bus_station()
    bus_station()
  elif "c" in answer:
    # else if player typed "car hire" or "c" lead him to car_hire()
    car_hire() 
  else:
    # else call game_over() function with the "reason" argument
    game_over("Typed in the wrong way")


# start the game
start()
```
